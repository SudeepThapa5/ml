EDA

```python
age_mean = df.Age.mean()
df["Age"].fillna(age_mean , inplace= True)
```

```python
dop_median = df["DurationOfPitch"].median()
df["DurationOfPitch"].fillna(dop_median, inplace = True)
```

```python
nof_median = df["NumberOfFollowups"].median()
df["NumberOfFollowups"].fillna(nof_median, inplace = True)
```

```python
nocv_median = df["NumberOfChildrenVisiting"].median()
nocv_median
```

```python
df["NumberOfChildrenVisiting"].fillna(nocv_median, inplace = True)
```

```python
mi_median = df["MonthlyIncome"].median()
mi_median
```

```python
df["MonthlyIncome"].fillna(mi_median, inplace = True)
```

```python
df["PreferredPropertyStar"].dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df["TypeofContact"].dtype
```

```python
df.dropna(subset=["PreferredPropertyStar"], inplace=True)
```

```python
df.TypeofContact.value_counts()
```

```python
df.isnull().sum()
```

```python
df.drop(["CustomerID","TypeofContact"], axis=1, inplace=True, errors='ignore')
```

```python
df.columns
```

```python
not_mean = df["NumberOfTrips"].mean()
not_mean
df["NumberOfTrips"].fillna(not_mean, inplace = True)
```

```python
df.isnull().sum()
```

Categorical values into Boolean

```python
occupation  = pd.get_dummies(df['Occupation'], drop_first=True)
gender = pd.get_dummies(df['Gender'], drop_first=True)
marital = pd.get_dummies(df['MaritalStatus'], drop_first=True)
designation = pd.get_dummies(df['Designation'], drop_first=True)
```

```python
df.drop(['Occupation','Gender','MaritalStatus','Designation'],axis=1,inplace=True)
```

```python
df.drop(["ProductPitched"], axis = 1, inplace=True)
```

```python
df.columns
```

```python
df = pd.concat([df, occupation, gender, marital, designation],axis=1)
```

```python
x=df.drop('ProdTaken',axis=1)
y=df['ProdTaken']
```

Training The Model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classifier=DecisionTreeClassifier()
```

```python
x_train.dtypes
```

```python
df["PreferredPropertyStar"] = pd.to_numeric(df["PreferredPropertyStar"], errors="coerce")
df.dropna(subset=["PreferredPropertyStar"], inplace=True)
```

```python
tree_classifier.fit(x_train,y_train)
```

Decision Tree

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier,filled=True)
```

```python
prediction=tree_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,prediction)
print(cm)
```

```python
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds")
```

```python
print(classification_report(y_test,prediction))
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf = RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(x_train, y_train)
```

```python
rf_prediction=rf.predict(x_test)
rf_prediction
```

```python
accurate=accuracy_score(y_test,rf_prediction)*100
print("The accuracy score is:",accurate,"%")
```

```python
rf_cm=confusion_matrix(y_test,rf_prediction)
rf_cm
```

```python
sns.heatmap(rf_cm, annot=True, fmt="d", cmap="Reds")
```

```python
print(classification_report(y_test,rf_prediction))
```
