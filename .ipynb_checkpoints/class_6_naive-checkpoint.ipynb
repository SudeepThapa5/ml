```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape,x_test.shape
```

```python
categorical = [col for col in x_train.columns if x_train[col].dtypes=='O']
categorical
```

```python
numerical = [col for col in x_train.columns if x_train[col].dtypes!='O']
numerical
```

```python
x_train[categorical].isnull().sum()
```

```python
work_class_mode= x_train["workclass"].mode()[0]
occupation_mode= x_train["occupation"].mode()[0]
native_country_mode= x_train["native_country"].mode()[0]
for d in [x_train,x_test]:
    d['workclass'].fillna(work_class_mode,inplace=True)
    d['occupation'].fillna(occupation_mode,inplace=True)
    d['native_country'].fillna(native_country_mode,inplace=True)
```
