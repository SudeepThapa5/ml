```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
df
```

```python
df.keys()
```

```python
cols = df["feature_names"]
```

```python
x = pd.DataFrame(df.data, columns = cols)
y = pd.Series(df["target"])
```

```python
x
```

```python
y
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 15)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn import tree
plt.figure(figsize = (12,8))
tree.plot_tree(dtc, filled = True)
```

```python
 params = {
     'criterion' : ['gini', 'entropy'],
     'max_depth': [2,3,4,5],
     'max_features' : ['sqrt', 'log2','auto']
 }
```

```python
grid = GridSearchCV(tree_model, param_grid = params, cv = 5, scoring = 'accuracy')
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```
